```python
def extract_unique_tags_with_ids(df, column_name):
  """
  Extracts unique tags from a comma-separated column in a Pandas DataFrame and assigns IDs.

  Args:
    df: The input DataFrame.
    column_name: The name of the column containing comma-separated tags.

  Returns:
    A dictionary mapping tag IDs to tags.
  """
  unique_tags = {}
  tag_id_counter = 1  # Start tag IDs from 1

  for _, row in df.iterrows():
    # Check if the value is a list (or list-like)
    if isinstance(row[column_name], (list, pd.Series)):
        # Iterate through the list of tags
        for tag in row[column_name]:
            cleaned_tag = str(tag).strip() # remove leading/trailing whitespace
            if cleaned_tag and cleaned_tag not in unique_tags.values():
                unique_tags[tag_id_counter] = cleaned_tag
                tag_id_counter += 1
    # If not a list, treat it as a single value
    elif pd.notna(row[column_name]):
        tags = str(row[column_name]).split(',')
        for tag in tags:
            cleaned_tag = tag.strip() # remove leading/trailing whitespace
            if cleaned_tag and cleaned_tag not in unique_tags.values():
                unique_tags[tag_id_counter] = cleaned_tag
                tag_id_counter += 1

  return unique_tags

columns_to_process = ['career_immediate', 'career_ultimate', 'reason_for_choosing_major', 'pros_major', 'lacking_lecture_areas', 'improvement_education', 'minor_detail']
all_unique_tags = {}  # Store all unique tags with IDs

for column in columns_to_process:
    unique_tags_with_ids = extract_unique_tags_with_ids(df, column)
    all_unique_tags[column] = unique_tags_with_ids # Store for later use
    print(f"Unique tags for column '{column}':")
    for tag_id, tag in unique_tags_with_ids.items():
        print(f"ID: {tag_id}, Tag: {tag}")
    print("-" * 20)

print("\nAll Unique Tags with IDs:")
print(all_unique_tags) # Print the dictionary containing all unique tags and IDs
```

```python
# minor_status

def extract_unique_single_select_tags(df, column_names):
    """
    Extracts unique tags from single-select columns and assigns IDs.

    Args:
        df: The input DataFrame.
        column_names: A list of column names to process.

    Returns:
        A dictionary where keys are column names and values are dictionaries
        mapping tag IDs to tags for each column.
    """

    all_unique_tags = {}
    for column in column_names:
        unique_tags = {}
        tag_id_counter = 1

        for tag in df[column].dropna().unique():  # Iterate over unique non-null values
            cleaned_tag = str(tag).strip()
            if cleaned_tag not in unique_tags.values():
                unique_tags[tag_id_counter] = cleaned_tag
                tag_id_counter += 1
        all_unique_tags[column] = unique_tags
    return all_unique_tags

columns_to_process = ['admission_method', 'minor_status']
single_select_tags = extract_unique_single_select_tags(df, columns_to_process)

single_select_tags
```

```python
import plotly.graph_objects as go

def visualize_tag_changes(df, all_unique_tags):
    reason_tags = all_unique_tags['reason_for_choosing_major']
    pros_tags = all_unique_tags['pros_major']

    # Find common tags between reason and pros
    common_tags = set(reason_tags.values()) & set(pros_tags.values())

    fig = go.Figure()

    # Visualize changes for each common tag
    for tag in common_tags:
        reason_count = df['reason_for_choosing_major'].str.contains(tag, na=False).sum()
        pros_count = df['pros_major'].str.contains(tag, na=False).sum()
        fig.add_trace(go.Scatter(x=['reason_for_choosing_major', 'pros_major'],
                                 y=[reason_count, pros_count],
                                 mode='lines+markers',
                                 name=tag))

    fig.update_layout(title='Change in Tag Frequency: Reason for Choosing Major to Pros of Major',
                      xaxis_title='Stage',
                      yaxis_title='Frequency')
    fig.show()


visualize_tag_changes(df, all_unique_tags)
```
